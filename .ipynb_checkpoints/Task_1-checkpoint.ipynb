#Task 1: Conceptual Understanding

Q1: What is the difference between "Love" and "love" in NLP?
Answer: In Natural Language Processing, "Love" and "love" are treated as different tokens because NLP models are case-sensitive by default. This means the model considers them as separate words, even though they carry the same meaning. This increases the vocabulary size unnecessarily and can reduce model efficiency. To avoid this issue, text is usually converted to lowercase during preprocessing so that similar words are treated as the same token.

Q2: What happens if stopwords are not removed?
Answer: If stopwords are not removed, the text data will contain many common words that do not add significant meaning to the sentence. This increases noise in the data and can make it harder for machine learning models to identify important patterns. It also increases computational complexity because the model has to process more unnecessary tokens. As a result, model performance and efficiency may decrease. However, in some cases such as sentiment analysis, removing stopwords may remove important context (e.g., "not"), which can change the meaning of a sentence.

Q3: Provide two real-world scenarios where removing stopwords can be harmful.
Answer: Removing stopwords can be harmful in tasks where sentence meaning depends on small but important words.

One scenario is sentiment analysis, where words like "not" are crucial. For example, the sentence "I do not like this product" would become "I like this product" if stopwords are removed, completely reversing the sentiment.

Another scenario is question answering or search systems, where stopwords help define intent. For example, removing words from "What is the best way to learn Python?" may distort the structure and meaning of the question, making it harder for the system to understand user intent correctly.

Q4: Explain the difference between stemming and lemmatization.
Answer: Stemming and lemmatization are both techniques used to reduce words to their base or root form in NLP.

Stemming is a simple and fast method that removes suffixes from words without considering their meaning. As a result, it may produce words that are not linguistically correct, such as "studies" becoming "studi".

Lemmatization, on the other hand, is a more advanced technique that uses vocabulary and morphological analysis to return the correct base form of a word. For example, "studies" becomes "study" and "running" becomes "run".

Overall, stemming is faster but less accurate, while lemmatization is slower but produces more meaningful and correct words.

#TASK 2

```python
import re
from collections import Counter

def preprocess_text(text):
    
    # 1. Handle edge cases
    if not text or not isinstance(text, str):
        return [], ""
    
    # 2. Remove URLs
    text = re.sub(r'http\S+|www\S+', '', text)
    
    # 3. Remove emails
    text = re.sub(r'\S+@\S+', '', text)
    
    # 4. Remove numbers
    text = re.sub(r'\d+', '', text)
    
    # 5. Convert to lowercase
    text = text.lower()

    # 6. handle repeated characters
    text = re.sub(r'(.)\1{2,}', r'\1', text)

    # 7. remove special characters & emojis
    text = re.sub(r'[^a-z\s]', '', text)

    #8.  reove extra spaces
    text = re.sub(r'\s+', ' ', text).strip()
    
    # 9. tokenize
    tokens = text.split()

    # handle empty after cleaning
    if not tokens:
        return [], ""
    
    # remove short words but keep 'no', 'not'
    tokens = [word for word in tokens if len(word) > 2 or word in ['no', 'not']]
    
    clean_sentence = " ".join(tokens)
    
    return tokens, clean_sentence

print(preprocess_text("I am not happy with this"))
```

#TASK 3

```python
test_sentences = [
    "Get 100% FREE access now!!!",
    "I absolutely looooved this product 😍😍",
    "Worst service ever... 0/10",
    "Call me at 9876543210",
    "This is THE best course!!!",
    "Visit https://openai.com now!",
    "Nooooo this is baaad!!!",
    "OK OK OK I got it",
    "Win $$$ now!!! Limited offer!!!",
    "I am not happy with this"
]

for sentence in test_sentences:
    tokens, clean_sentence = preprocess_text(sentence)
    
    print("Original:", sentence)
    print("Tokens:", tokens)
    print("Cleaned Sentence:", clean_sentence)
    print("-" * 50)
```

## Task 4: Token Analytics

```python
for sentence in test_sentences:
    tokens, clean_sentence = preprocess_text(sentence)
    
    total_tokens = len(tokens)
    unique_tokens = len(set(tokens))
    
    if total_tokens > 0:
        avg_length = sum(len(word) for word in tokens) / total_tokens
    else:
        avg_length = 0
    
    print(f"Original Text: {sentence}")
    print(f"Tokens: {tokens}")
    print(f"Total Tokens: {total_tokens}")
    print(f"Unique Tokens: {unique_tokens}")
    print(f"Average Token Length: {avg_length:.2f}")
    print("=" * 60)
```

## Task 5: Frequency Analysis

```python
from collections import Counter

all_tokens = []

for sentence in test_sentences:
    tokens, _ = preprocess_text(sentence)
    all_tokens.extend(tokens)

word_freq = Counter(all_tokens)

top_10 = word_freq.most_common(10)

print("Top 10 Most Frequent Words:")
for word, count in top_10:
    print(f"{word}: {count}")

least_5 = word_freq.most_common()[:-6:-1]

print("\nTop 5 Least Frequent Words:")
for word, count in least_5:
    print(f"{word}: {count}")
```

## Task 6: Build Full Pipeline

```python
def full_pipeline(text_list):
    
    all_tokens = []
    clean_sentences = []
    
    for text in text_list:
        tokens, clean_sentence = preprocess_text(text)
        
        all_tokens.extend(tokens)
        clean_sentences.append(clean_sentence)
    
    return {
        "tokens": all_tokens,
        "clean_sentences": clean_sentences
    }

result = full_pipeline(test_sentences)

print("All Tokens:")
print(result["tokens"])

print("\nClean Sentences:")
for sentence in result["clean_sentences"]:
    print(sentence)
```
