# BERT Fine-Tuning for Text Classification
### NLP Assignment Submission

**Objective:** Fine-tune BERT on a Kaggle dataset and evaluate performance.

---

## 1. Install & Import Libraries

```
!pip install transformers datasets scikit-learn torch

import pandas as pd
import numpy as np
import torch
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, precision_recall_fscore_support, confusion_matrix
from transformers import AutoTokenizer, AutoModelForSequenceClassification, Trainer, TrainingArguments
```

## 2. Load Dataset

```
df = pd.read_csv("IMDB Dataset.csv")

df = df.rename(columns={"review": "text", "sentiment": "label"})
df["label"] = df["label"].map({"positive": 1, "negative": 0})

df.head()
```

## 3. Data Preprocessing

```
df.dropna(inplace=True)

df["text"] = df["text"].str.lower()
df["text"] = df["text"].str.replace(r"[^a-zA-Z0-9\s]", "", regex=True)
```

## 4. Train / Validation / Test Split

```
train_texts, temp_texts, train_labels, temp_labels = train_test_split(
    df["text"], df["label"], test_size=0.3, random_state=42
)

val_texts, test_texts, val_labels, test_labels = train_test_split(
    temp_texts, temp_labels, test_size=0.5, random_state=42
)
```

## 5. Tokenization

```
tokenizer = AutoTokenizer.from_pretrained("bert-base-uncased")

def tokenize(texts):
    return tokenizer(list(texts), padding=True, truncation=True, max_length=256)

train_encodings = tokenize(train_texts)
val_encodings = tokenize(val_texts)
test_encodings = tokenize(test_texts)
```

## 6. Dataset Class

```
class Dataset(torch.utils.data.Dataset):
    def __init__(self, encodings, labels):
        self.encodings = encodings
        self.labels = list(labels)

    def __getitem__(self, idx):
        item = {key: torch.tensor(val[idx]) for key, val in self.encodings.items()}
        item["labels"] = torch.tensor(self.labels[idx])
        return item

    def __len__(self):
        return len(self.labels)

train_dataset = Dataset(train_encodings, train_labels)
val_dataset = Dataset(val_encodings, val_labels)
test_dataset = Dataset(test_encodings, test_labels)
```

## 7. Load Model

```
model = AutoModelForSequenceClassification.from_pretrained(
    "bert-base-uncased", num_labels=2
)
```

## 8. Metrics

```
def compute_metrics(pred):
    logits, labels = pred
    preds = np.argmax(logits, axis=1)

    precision, recall, f1, _ = precision_recall_fscore_support(labels, preds, average="binary")
    acc = accuracy_score(labels, preds)

    return {"accuracy": acc, "f1": f1, "precision": precision, "recall": recall}
```

## 9. Training Arguments

```
training_args = TrainingArguments(
    output_dir="./results",
    learning_rate=2e-5,
    per_device_train_batch_size=8,
    per_device_eval_batch_size=8,
    num_train_epochs=2,
    evaluation_strategy="epoch",
    save_strategy="epoch",
    logging_dir="./logs",
)
```

## 10. Trainer

```
trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=train_dataset,
    eval_dataset=val_dataset,
    tokenizer=tokenizer,
    compute_metrics=compute_metrics,
)
```

## 11. Training

```
trainer.train()
```

## 12. Evaluation

```
preds_output = trainer.predict(test_dataset)

preds = np.argmax(preds_output.predictions, axis=1)
labels = preds_output.label_ids

print("Accuracy:", accuracy_score(labels, preds))

precision, recall, f1, _ = precision_recall_fscore_support(labels, preds, average="binary")
print("Precision:", precision)
print("Recall:", recall)
print("F1 Score:", f1)

print("Confusion Matrix:\n", confusion_matrix(labels, preds))
```

## 13. Experiments

```
# Freeze BERT layers
for param in model.bert.parameters():
    param.requires_grad = False
```

```
# Fine-tune last 2 layers
for name, param in model.bert.named_parameters():
    if "encoder.layer.10" in name or "encoder.layer.11" in name:
        param.requires_grad = True
    else:
        param.requires_grad = False
```
